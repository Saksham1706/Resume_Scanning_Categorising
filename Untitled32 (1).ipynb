```python
import os

# Reuse your existing folder variables from the notebook
ENG_DIR = '/content/drive/MyDrive/proj_resume/ENGINEERING'
IT_DIR  = '/content/drive/MyDrive/proj_resume/INFORMATION-TECHNOLOGY'

def canon_cat(s):
    s = str(s).strip().upper()
    if s in {'IT','INFORMATION TECHNOLOGY','INFO TECH','INFORMATION-TECHNOLOGY'}:
        return 'INFORMATION-TECHNOLOGY'
    if s in {'ENGINEER','ENGINEERING'}:
        return 'ENGINEERING'
    return s

df['Category_canon'] = df['Category'].apply(canon_cat)

def make_path(row):
    rid = str(row['ID']).strip()  # Column is 'ID' per your CSV preview
    if row['Category_canon'] == 'ENGINEERING':
        return f"{ENG_DIR}/{rid}.pdf"
    if row['Category_canon'] == 'INFORMATION-TECHNOLOGY':
        return f"{IT_DIR}/{rid}.pdf"
    return None

df['pdf_path'] = df.apply(make_path, axis=1)
df['exists'] = df['pdf_path'].apply(lambda p: isinstance(p, str) and os.path.isfile(p))
df = df[df['exists']].drop(columns=['exists']).copy()
print('Rows with valid pdf_path:', len(df))
```

```python
# Parse PDFs to text (PyMuPDF preferred, PyPDF as fallback)
!pip -q install pymupdf pypdf

import fitz  # PyMuPDF
from PyPDF2 import PdfReader

def pdf_to_text(path):
    # 1) PyMuPDF
    try:
        with fitz.open(path) as doc:
            pages = [page.get_text("text") for page in doc]
        txt = "\n".join(pages)
        if txt and txt.strip():
            return txt
    except Exception:
        pass
    # 2) PyPDF2 fallback
    try:
        reader = PdfReader(path)
        pages = [page.extract_text() or "" for page in reader.pages]
        return "\n".join(pages)
    except Exception:
        return ""

df['raw_text'] = df['pdf_path'].apply(pdf_to_text)
df = df[df['raw_text'].str.strip().ne("")].copy()
print('Parsed non-empty texts:', len(df))
```

```python
# Clean text and split into chunks (token-aware via tiktoken; fallback to length-based if needed)
%pip install --quiet tiktoken

import re, math, json
import tiktoken

# 1) Clean
def clean_text(t):
    t = t.replace('\u00a0', ' ')
    t = re.sub(r'[ \t]+', ' ', t)
    t = re.sub(r'\n{3,}', '\n\n', t)
    return t.strip()

df['text_clean'] = df['raw_text'].apply(clean_text)

# 2) Simple section heuristic to keep semantics
SECTION_HEADERS = [
    r'(?:SUMMARY|OBJECTIVE|PROFILE)',
    r'(?:EXPERIENCE|WORK EXPERIENCE|PROFESSIONAL EXPERIENCE|EMPLOYMENT)',
    r'(?:EDUCATION|QUALIFICATIONS|ACADEMIC)',
    r'(?:SKILLS|TECHNICAL SKILLS|CORE SKILLS)',
    r'(?:PROJECTS|PERSONAL PROJECTS|ACADEMIC PROJECTS)',
    r'(?:CERTIFICATIONS|LICENSES)',
    r'(?:PUBLICATIONS)',
    r'(?:ACHIEVEMENTS|AWARDS)',
]
header_regex = re.compile(rf"^({'|'.join(SECTION_HEADERS)})\b", re.IGNORECASE)

def sectionize(text):
    lines = [l.strip() for l in text.splitlines()]
    secs, cur, buf = [], 'UNKNOWN', []
    for ln in lines:
        m = header_regex.match(ln)
        if m:
            if buf:
                secs.append((cur, "\n".join(buf).strip()))
            cur = m.group(1).upper()
            buf = []
        else:
            buf.append(ln)
    if buf:
        secs.append((cur, "\n".join(buf).strip()))
    return secs or [('FULL', text)]

df['sections'] = df['text_clean'].apply(sectionize)

# 3) Token-aware chunking with overlap
enc = tiktoken.get_encoding("cl100k_base")  # good default tokenizer
CHUNK_TOKENS = 220
OVERLAP_TOKENS = 40

def chunk_text(txt, chunk_tokens=CHUNK_TOKENS, overlap_tokens=OVERLAP_TOKENS):
    toks = enc.encode(txt)
    chunks = []
    i = 0
    while i < len(toks):
        j = min(i + chunk_tokens, len(toks))
        chunk = enc.decode(toks[i:j])
        chunks.append(chunk.strip())
        if j == len(toks): break
        i = j - overlap_tokens
    return [c for c in chunks if c]

# Build a flat list of chunks with metadata
records = []
for _, row in df.iterrows():
    rid = str(row['ID'])
    cat = row.get('Category', '')
    pdf = row.get('pdf_path', '')
    for (sec_name, sec_text) in row['sections']:
        for idx, ch in enumerate(chunk_text(sec_text)):
            records.append({
                'id': rid,
                'category': cat,
                'section': sec_name,
                'chunk_id': f"{rid}:{sec_name}:{idx}",
                'pdf_path': pdf,
                'text': ch
            })

print("Total chunks:", len(records))

# Save chunks JSONL for embedding + vector DB step
import os
OUT_DIR = '/content/drive/MyDrive/proj_resume/_chunks'
os.makedirs(OUT_DIR, exist_ok=True)
OUT_JSONL = os.path.join(OUT_DIR, 'resumes_chunks.jsonl')
with open(OUT_JSONL, 'w', encoding='utf-8') as f:
    for r in records:
        f.write(json.dumps(r, ensure_ascii=False) + '\n')
print("Wrote chunks to:", OUT_JSONL)
```

```python
%pip install --quiet faiss-cpu sentence-transformers

import json, os
import numpy as np
from tqdm.auto import tqdm
from sentence_transformers import SentenceTransformer
import faiss

# Paths
CHUNKS_JSONL = '/content/drive/MyDrive/proj_resume/_chunks/resumes_chunks.jsonl'
INDEX_DIR = '/content/drive/MyDrive/proj_resume/_index'
os.makedirs(INDEX_DIR, exist_ok=True)
EMB_NPY = os.path.join(INDEX_DIR, 'embeddings.npy')
META_JSONL = os.path.join(INDEX_DIR, 'metadata.jsonl')
FAISS_IDX = os.path.join(INDEX_DIR, 'faiss.index')

# Load chunks
texts, metas = [], []
with open(CHUNKS_JSONL, 'r', encoding='utf-8') as f:
    for line in f:
        r = json.loads(line)
        texts.append(r['text'])
        metas.append({
            'id': r['id'],
            'category': r.get('category',''),
            'section': r.get('section',''),
            'chunk_id': r.get('chunk_id',''),
            'pdf_path': r.get('pdf_path','')
        })

print('Loaded chunks:', len(texts))

# Load embedding model (good general-purpose)
model = SentenceTransformer('intfloat/e5-base-v2')  # or 'BAAI/bge-base-en-v1.5'
# E5 expects "query: ..." / "passage: ..." formats; for corpus use passage:
corpus_inputs = [f"passage: {t}" for t in texts]

# Embed in batches
embs = model.encode(corpus_inputs, batch_size=64, show_progress_bar=True, normalize_embeddings=True)
embs = np.asarray(embs, dtype='float32')
print('Emb shape:', embs.shape)

# Save embeddings and metadata
np.save(EMB_NPY, embs)
with open(META_JSONL, 'w', encoding='utf-8') as f:
    for m in metas:
        f.write(json.dumps(m, ensure_ascii=False) + '\n')

# Build FAISS index (cosine via Inner Product with normalized vectors)
d = embs.shape[1]
index = faiss.IndexFlatIP(d)
index.add(embs)
faiss.write_index(index, FAISS_IDX)
print('FAISS index saved at:', FAISS_IDX)
print('Metadata saved at:', META_JSONL)
print('Embeddings saved at:', EMB_NPY)
```

```python
import os, json
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss

INDEX_DIR = '/content/drive/MyDrive/proj_resume/_index'
FAISS_IDX = os.path.join(INDEX_DIR, 'faiss.index')
EMB_NPY   = os.path.join(INDEX_DIR, 'embeddings.npy')  # not needed for search but kept for reference
META_JSONL= os.path.join(INDEX_DIR, 'metadata.jsonl')

# Load index and metadata
index = faiss.read_index(FAISS_IDX)
metas = [json.loads(l) for l in open(META_JSONL, 'r', encoding='utf-8').read().splitlines()]

# Load same embedding model used for corpus (E5-base-v2)
model = SentenceTransformer('intfloat/e5-base-v2')

def search(query, k=5, category_filter=None):
    # E5 format: "query: ..."
    qvec = model.encode([f"query: {query}"], normalize_embeddings=True).astype('float32')
    D, I = index.search(qvec, k*5)  # retrieve extra for filtering
    hits = []
    for idx, score in zip(I[0], D[0]):
        if idx == -1:
            continue
        m = metas[idx]
        if category_filter and m.get('category') not in category_filter:
            continue
        hits.append({'score': float(score), **m})
        if len(hits) == k:
            break
    return hits

# Example queries to validate retrieval quality
queries = [
    "3+ years experience in Python and SQL for data engineering",
    "React and Node.js full-stack web development projects",
    "Embedded systems C/C++ with microcontrollers",
    "Machine learning with PyTorch and NLP experience",
]

for q in queries:
    res = search(q, k=5, category_filter=None)  # or {'ENGINEERING'} / {'INFORMATION-TECHNOLOGY'}
    print("\nQ:", q)
    for r in res:
        print(f"  score={r['score']:.3f}  id={r['id']}  cat={r['category']}  section={r['section']}  chunk_id={r['chunk_id']}")
```

```python
# Upload one PDF resume -> parse -> chunk (220/40) -> embed with same model -> append to existing FAISS index
# Also auto-assign category: ENGINEERING vs INFORMATION-TECHNOLOGY via simple keyword heuristic.

import os, re, json, uuid
import numpy as np
from google.colab import files
import fitz  # PyMuPDF
from PyPDF2 import PdfReader
import tiktoken
from sentence_transformers import SentenceTransformer
import faiss

# Reuse your existing paths/index
INDEX_DIR = '/content/drive/MyDrive/proj_resume/_index'
FAISS_IDX = os.path.join(INDEX_DIR, 'faiss.index')
META_JSONL = os.path.join(INDEX_DIR, 'metadata.jsonl')

# 1) Upload
print("Upload your PDF resume...")
uploaded = files.upload()
assert uploaded, "No file uploaded."
pdf_local = list(uploaded.keys())[0]
assert pdf_local.lower().endswith('.pdf'), "Please upload a .pdf file."

# 2) Extract text
def pdf_to_text(path):
    try:
        with fitz.open(path) as doc:
            pages = [page.get_text("text") for page in doc]
        txt = "\n".join(pages)
        if txt and txt.strip():
            return txt
    except Exception:
        pass
    try:
        reader = PdfReader(path)
        pages = [page.extract_text() or "" for page in reader.pages]
        return "\n".join(pages)
    except Exception:
        return ""

raw_text = pdf_to_text(pdf_local).strip()
assert raw_text, "Could not extract text from the uploaded PDF."

# 3) Clean + chunk identical to your pipeline (220 tokens, 40 overlap)
def clean_text(t):
    t = t.replace('\u00a0', ' ')
    t = re.sub(r'[ \t]+', ' ', t)
    t = re.sub(r'\n{3,}', '\n\n', t)
    return t.strip()

text_clean = clean_text(raw_text)
enc = tiktoken.get_encoding("cl100k_base")
CHUNK_TOKENS, OVERLAP_TOKENS = 220, 40

def chunk_text(txt, chunk_tokens=CHUNK_TOKENS, overlap_tokens=OVERLAP_TOKENS):
    toks = enc.encode(txt)
    chunks = []
    i = 0
    while i < len(toks):
        j = min(i + chunk_tokens, len(toks))
        chunk = enc.decode(toks[i:j]).strip()
        if chunk:
            chunks.append(chunk)
        if j == len(toks): break
        i = j - overlap_tokens
    return chunks

chunks = chunk_text(text_clean)
assert chunks, "No chunks produced from text."

# 4) Simple category guess (overwrite or adjust as you like)
it_keywords = [
    'software', 'developer', 'developer.', 'java', 'python', 'sql', 'javascript',
    'web', 'full stack', 'frontend', 'backend', 'cloud', 'aws', 'gcp', 'azure',
    'data engineer', 'data scientist', 'ml', 'ai', 'nlp', 'devops', 'docker', 'kubernetes',
    'react', 'node', 'django', 'flask', 'spring', 'microservices', 'rest api'
]
eng_keywords = [
    'mechanical', 'manufacturing', 'cad', 'solidworks', 'ansys', 'catia',
    'automobile', 'automotive', 'thermodynamics', 'fluid', 'fea', 'cfd',
    'electrical', 'electronics', 'embedded', 'pcb', 'matlab', 'simulink',
    'civil', 'structural', 'construction', 'survey', 'design engineer'
]

txt_low = text_clean.lower()
it_hits = sum(kw in txt_low for kw in it_keywords)
eng_hits = sum(kw in txt_low for kw in eng_keywords)
category = 'INFORMATION-TECHNOLOGY' if it_hits >= eng_hits else 'ENGINEERING'

# 5) Prepare metadata for each chunk
doc_id = str(uuid.uuid4())[:8]
metas_new = [{
    'id': doc_id,
    'category': category,
    'section': 'FULL',
    'chunk_id': f"{doc_id}:FULL:{i}",
    'pdf_path': os.path.abspath(pdf_local)
} for i in range(len(chunks))]

# 6) Embed with the same model you used (E5-base-v2, cosine via IP)
model = SentenceTransformer('intfloat/e5-base-v2')
corpus_inputs = [f"passage: {t}" for t in chunks]
embs_new = model.encode(corpus_inputs, batch_size=64, normalize_embeddings=True).astype('float32')
print("New embeddings shape:", embs_new.shape, "| Category:", category)

# 7) Append to existing FAISS index + metadata
assert os.path.isfile(FAISS_IDX), "FAISS index not found; build the main index first."
index = faiss.read_index(FAISS_IDX)
assert index.d == embs_new.shape[1], f"Dim mismatch: index {index.d} vs embeddings {embs_new.shape[1]}"
index.add(embs_new)
faiss.write_index(index, FAISS_IDX)

with open(META_JSONL, 'a', encoding='utf-8') as f:
    for m in metas_new:
        f.write(json.dumps(m, ensure_ascii=False) + '\n')

print(f"Appended {len(chunks)} vectors and metadata for document {doc_id}. Auto-category: {category}")
```
